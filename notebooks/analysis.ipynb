## Data Loading

```python
import pandas as pd

# Load cleaned data
df = pd.read_excel('../data/cleaned_data.xlsx')

df.head()
```

The cleaned dataset was loaded for analysis. It contains sales, customer, and operational data for Café Harmony.

## Data Cleaning

```python
# Check structure
df.info()

# Check missing values
df.isnull().sum()
```

```python
# Fill missing values
df = df.fillna(0)
```

Basic data checks were performed, including data types and missing values. Missing values were handled to ensure consistency for analysis.

## Exploratory Analysis

```python
total_sales = df['Total_Sale'].sum()
total_sales
```

```python
df.groupby('Product')['Total_Sale'].sum().sort_values(ascending=False)
```

## Visualizations

```python
import matplotlib.pyplot as plt
```

```python
df.groupby('Product')['Total_Sale'].sum().plot(kind='bar')
plt.title('Sales by Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
df['Sale_Date'] = pd.to_datetime(df['Sale_Date'])

monthly_sales = df.groupby(pd.Grouper(key='Sale_Date', freq='ME'))['Total_Sale'].sum()

monthly_sales.plot(figsize=(12,5), marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

## Sales Trend Analysis

Daily sales data was aggregated to a monthly level to reduce noise and better visualize overall trends and patterns.

This provides a clearer understanding of business performance over time.

```python
df.groupby('Location')['Total_Sale'].sum().plot(kind='bar')
plt.title('Sales by Location')
plt.show()
```

## Insights

- Certain products Muffin, Cappuccino, nd Ice Tea) generate the highest revenue
- Sales trends vary over time, indicating peak periods
- City's locational differences highlight performance variation across locations
